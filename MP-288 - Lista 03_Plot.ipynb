# Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Variáveis numéricas

```python
# variaveis para minimização
x = np.linspace(0, np.pi/2, 100)
y = np.linspace(15, 25, 100)
beta, d = np.meshgrid(x, y)

# constantes 
dr = 12
f = 1.5
mc = 600
mm = 59
mq_bar = 1150
ro = 2.7
z = 3
A = 1.4
```

# Funções

```python
Iyy = 4*((d**2*mm+d**3*ro*A/3)*(np.sin(beta))**2 + z**2*(mm + d*ro*A)) # função objetivo 
g1 = dr/2 + f - d*np.cos(beta)                                         # restrição de ângulo por baixo
g2 = dr/2 + f - d*np.sin(beta)                                         # restrição de ângulo por cima
g3 = mc + 4*(mm+d*ro*A) - mq_bar                                       # restrição de massa
```

# Plot

```python
fig = plt.figure()
x = np.arange(0.8, 1.2, 0.01)
y1 = (dr/2 + f)/np.cos(x)
y2 = 20.8
plt.grid(b=True, which='major', color='black', linestyle='--');
contours = plt.contour(beta, d, Iyy, 11)                            # plot função objetivo
plt.clabel(contours, inline=True, fontsize=10)
plt.contour(beta, d, g1, 0, colors = 'red')                         # plot restrição angulo por baixo
plt.contour(beta, d, g2, 0, colors = 'magenta')                     # plot restrição angulo por cima
plt.contour(beta, d, g3, 0, colors = 'blue')                        # plot restrição de massa
plt.text(0.53, 15.2, '1',fontsize = 15)                             # plot ponto de verificação 1
plt.text(0.40, 21.0, '2',fontsize = 15)                             # plot ponto de verificação 2
plt.text(0.43, 18.0, '3',fontsize = 15)                             # plot ponto de verificação 3
plt.fill_between(x, y1, y2, color = 'yellow', alpha = 0.5)          # marcação de amarelo da área entre as curvas
x = np.arange(0.38, 0.81, 0.01)
y1 = (dr/2 + f)/np.sin(x)
y2 = 20.8
plt.fill_between(x, y1, y2, color = 'yellow', alpha = 0.5)          # marcação de amarelo da área entre as curvas
plt.plot(0.52,15,color='black',marker='o',markersize=8)
plt.plot(0.37,20.8,color='black',marker='o',markersize=8)
plt.ylabel('d [cm]')
plt.xlabel(r'$\beta$')
plt.ylim([15,25])
fig.savefig('multicopter.png',dpi=300)                              # salvar figura como png (dpi = número de pixels)
plt.show()
```
