````python
def f(x):
    f = x[0]**2 + x[1]**2 - 3*x[0]*x[1]
    return f

def g(x):
    g = x[0]**2/6 + x[1]**2/6 - 1
    return g

def h(x):
    h = -x[0]
    return h

def k(x):
    k = -x[1]
    return k
````
